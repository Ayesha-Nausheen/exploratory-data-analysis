- total bill: total bill at restaurant
- tip: tip given to the waiter
- sex: gender of that person (who paid the bill)
- smoker: whether the person is smoker or non-smoker
- day: which day they came to restaurant
- time: which time they came to restaurant
- size: total no.of members came to restaurant

# Joint Plot
A join plot allows to stude the relationship between two numeric variables. the central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D desity plot

```python
sns.jointplot(x='total_bill',y='tip',data=df)
plt.show()
```

# Violin Plot
A violin plot helps to see both the distribution of dara in terms of kernal density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
plt.show()
```

# Pair Plot - Multiple Continious Variables
A pairs plot is known as a scatterplot in which one variable in the same data row is matched with variables value, like this: pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df,vars=continous)
plt.show()
```

# Heat Map
- A heatmap uses colored cells to represent relation between variables

#### Heatmap (for correlation)
- A correlation heatmap uses colored cells to show a 2D correlation matrix (table) between two numeric dimensions.
- It is very important in Feature Selection
- Correlation (r) = COV(x.y)/sigma x.sigma y

```python
c_m=df[['total_bill','tip']].corr()
c_m
```

```python
sns.heatmap(c_m,annot=True)
```
